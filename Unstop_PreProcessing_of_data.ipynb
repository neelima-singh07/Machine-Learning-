# Importing Libraries and Dataset


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("//content//Covid_Data_new.xlsx")
data
```

```python
X=data.iloc[:,:-1].values #returns array
Y=data.iloc[:,-1].values
print(X)
```

```python
data.isnull().sum()
```

# Handling Missing data

```python
from  sklearn.impute import SimpleImputer #import library
imputer=SimpleImputer(missing_values=np.nan,strategy='mean') #use simple imp to replace nan values by mean

# Apply imputer to the 'age' column (index 3 in the transformed X)
imputer.fit(X[:,3:4])
X[:,3:4]=imputer.transform(X[:,3:4])

# Apply imputer to the 'Blood O2 Level in Percentage' column (index 6 in the transformed X)
imputer.fit(X[:,6:7])
X[:,6:7]=imputer.transform(X[:,6:7])

print(X)
```

# Encoding Categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
#for independent variable
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1]),
                                     ('encoder2', OneHotEncoder(), [4, 5])],
                       remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
Y
```

# Seprating Training and Testing data

```python
from sklearn.model_selection import  train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,6:]=sc.fit_transform(X_train[:,6:])
X_test[:,6:]=sc.transform(X_test[:,6:])
```
