```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing dataset

```python
data=pd.read_excel('/content/regression sheet.xlsx')
data
```

# splitting dataset

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train)
print()
print(x_test)
```

```python
print(y_test)
```

# Feature scaling

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(x_train)
```

```python
print(y_test)
```

# Applying Logigistic regx


```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression (random_state=42)
classifier.fit(x_train,y_train)
```

# Predicting

```python
classifier.predict([[15,26678]])
```

```python
# Try predicting with different values for Age and Mobile Phone Price
new_prediction = classifier.predict([[40, 80000]]) # Example: Age 40, Mobile Phone Price 80000
print(new_prediction)
```

```python
# Predict with a younger age and lower mobile phone price
new_prediction_low_values = classifier.predict([[28, 1000]]) # Example: Age 20, Mobile Phone Price 10000
print(new_prediction_low_values)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=classifier.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)


```
